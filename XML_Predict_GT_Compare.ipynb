```python
import numpy as np
import os
from xml.etree import ElementTree
from xml.etree.ElementTree import XMLParser
from numpy import array
float32 = 'float32'
```

```python
def ground_true_answer(xml_file):
    tree = ElementTree.parse(xml_file)
    root = tree.getroot()
    symbols = ['♠','♥','♦️','♣','♦','♥','♠','♣'] 
    suite_list = []
    rang_list = []
    money_list = []    
    card_list = [] 
    for obj in root.findall('object'):
        for i in obj:
            if i.text == 'cards':
                suite =  (obj.findall('attributes')[0].text).split('=')[1]        
                a = (obj.findall('polygon')[0])            
                coord = []
                for i in a.findall('pt'):       
                    pair = []
                    for el in i:                
                        pair+=[int(float(el.text))]
                    coord +=[pair]
                if suite in symbols:
                    suite_list += [[suite, coord]]
                elif suite not in symbols:
                    rang_list += [[suite, coord]]

            elif i.text == 'money':
                money =  (obj.findall('attributes')[0].text).split('=')[1]
                money_list += [money]
      
    for i in suite_list:
        for j in rang_list:
            if int(i[1][0][0])-10 < int(j[1][0][0]) < int(i[1][0][0]) + 10\
                        and int(i[1][0][1])-50 < int(j[1][0][1]) < int(i[1][0][1]) + 50:
                card_list += [j[0]+i[0]]
            
    return card_list, money_list


def predict_answer(predict):    
    card_list = []
    money_list = []
    symbols = ['♠','♥','♦️','♣','♦','♥','♠','♣'] 
    for i in predict:
        if i[0].startswith('$'):
            money_list += [i[0]]
        elif len(i[0])==2 and i[0][1] in symbols:
            card_list += [i[0]]
        elif len(i[0])==3 and i[0][2] in symbols:
            card_list += [i[0]]
    return card_list, money_list


def accuracy(ground,predict):   
    count = 0
    for i in ground:
        if i in predict:
            count += 1
    precision =  (round(count*100/len(ground),2))  
    # метрика - оценка точности, должна быть 100% 
    count = 0
    for i in predict:
        if i in ground:
            count += 1
    recall = (100-round(count*100/len(predict),2)) 
    # метрика - оценка ложных срабатываний, должна быть 0%
    return [precision,recall]
```

```python
def answer_accuracy(ground_true_xml, predict_list):
    ground_true_data = ground_true_answer(ground_true_xml)
    predict_data = predict_answer(predict_list)
    card_accuracy = accuracy(ground_true_data[0],predict_data[0])
    money_accuracy = accuracy(ground_true_data[1], predict_data[1])
    return card_accuracy, money_accuracy
```
